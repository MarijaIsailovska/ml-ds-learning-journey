```python
df1=pd.read_csv("diabetes_data1.csv",index_col='PatientID')
df2=pd.read_csv("diabetes_data1_newRows.csv", index_col='PatientID')
df3=pd.read_csv("diabetes_data2.csv",index_col="PatientID")
```

# Joining of dataframes
# - Inner join
# - Left join
# - Right join
# - Outer join

# Joining is going to happen based on common primary key

```python
df_inner=pd.merge(concatenated_df,df3, on='PatientID',how='inner')
print(df_inner.shape)
df_inner
```

```python
df_left=pd.merge(concatenated_df,df3, on='PatientID',how='left')
print(df_left.shape)
df_left
```

```python
df_right=pd.merge(concatenated_df,df3, on='PatientID',how='right')
print(df_right.shape)
df_right
```

```python
df_outer=pd.merge(concatenated_df,df3, on='PatientID',how='outer')
print(df_outer.shape)
df_outer
```
