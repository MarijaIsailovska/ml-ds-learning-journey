# Seaborn
Seaborn is a powerful Python data visualization library built on top of Matplotlib. It provides a high-level interface for drawing attractive and informative statistical graphics. Seaborn makes it easier to create complex visualizations by providing simple functions for common tasks anc better aesthetics out of the box.
Seaborn works well with Pandas DataFrames and integrates seamlessly with Natplotlib. It provides functionalities for visualizing complex datasets and relationships in a more aesthetically pleasing way.
- Built-in Themes and Color Palettes: Provides aesthetically pleasing default styles and color palettes.
- Statistical Visualization: Includes functions to visualize complex statistical relationships such as linear regression, categorical data plots, and heatnaps.
- Integration with Pandas DataFrames: Works seamlessly with Pandas, allowing
easy plotting of DataFrane structures.
- Data Aggregation and Plotting: Automatically performs statistical aggregation and plotting in a single step.
- Extensive Customization: Supports extensive customization and fine-tuning of plot


```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
tips=sns.load_dataset("tips")
tips.head()
```

# Scatter Plot - used to display relationships between two continous variables 

```python
sns.scatterplot(x='total_bill',y='tip',data=tips,hue='day')
plt.title("Scatter Plot of Tip vs Total Bill")
plt.show()
```

# Line Plot - is used to display trends over time of continous variables 


```python
sns.lineplot(x='total_bill',y='tip',data=tips)
plt.title("Line Plot of Tip vs Total Bill")
plt.show()
```

# Histogram and KDE plot - shows distribution of a single continuous variable

```python
sns.histplot(tips['total_bill'],kde=True)
plt.title("Distribution")
plt.show()
```

# Box plot - Compare distributions across categories

```python
sns.boxplot(x='day',y='total_bill',data=tips)
plt.title("Box Plot of Tip vs Total Bill")
plt.show()
```

# Pair plot - Visualize pairwise relationship in a dataset

```python
sns.pairplot(tips,hue='day')
plt.show()
```

# Bar plot - Shows the distribution of a categorical variable 

```python
sns.barplot(x='day',y='total_bill',data=tips)
plt.title("Average Total Bill by day")
plt.show()
```

### Set style

```python
sns.set_style('whitegrid')
sns.scatterplot(x='total_bill', y='tip', data=tips)
plt.title('Scatter Plot with Whitegrid Style')
plt.show()
```

### Custom color

```python
sns.scatterplot(x='total_bill', y='tip', data=tips,hue='day',palette='coolwarm')
plt.title('Scatter Plot with custom color')
plt.show()
```

# Heatmap

```python
tips_numerical = tips[['total_bill','tip','size']]
```

```python
correlation_matrix=tips_numerical.corr()
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm')
plt.title('Correlation Heatmap')
plt.show()
```
