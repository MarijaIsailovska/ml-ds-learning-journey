# Data Cleaning and data transformation using Pandas
- Identify missing values and treating missing values in the dataset
- Identify duplicate rows and removing duplicate rows

```python
df.Insulin=df.Insulin.fillna(df.Insulin.mean())
df.isnull().sum()
```

```python
# Identify duplicate rows
df.duplicated()
```

```python
df=df.drop_duplicates()
df
```

```python
df.shape
```

# Outlier identification - IQR
# Outlier -> Lower than Q1 - 1.5*IQR or Higher than Q3 + 1.5*IQR

```python
df.Insulin
```

```python
Q1=df.Insulin.quantile(0.25)
Q3=df.Insulin.quantile(0.75)
IQR=Q3-Q1

lower_bound=Q1-1.5*IQR
upper_bound=Q3+1.5*IQR

print(lower_bound)
print(upper_bound)

df['is_outlier_insulin_col']=(df["Insulin"] < lower_bound)|(df["Insulin"] > upper_bound)
df
```

```python
mean_value=df.loc[~df['is_outlier_insulin_col'],"Insulin"].mean()
df.loc[df['is_outlier_insulin_col'],"Insulin"]=mean_value
```
