# ~ Matplotlib:
- Python Library used for creating visualization
- 2D Plotting
- Daa patterns, trends, insights
### Key features:
- Different types of plot types - line plot, bar charts, histogram, scatter plots, pie charts
- Customization - adjusting line style, adjust colors, labels, legends, annotations
- Integration with other libraries - numpy, pandas, scipy
- Support for interative visualization -
### Basic Framework of Matplotlib
- 1 Figure: This is the entire window or page that will contain the plots.
- 2 Axes: This is a single plot within a figure. A figure can contai multiple axes.
- 3 Artist: This is any element that can be rendered (e.g., Lines, text, labels).
### Key Components 
- Figure: plt.figure()
- Axes: fig.add_subplot() or plt.subplots()
- Plotting functions: plt.plot(), plt.scatter(), etc.

```python
import numpy as np
import pandas as pd 
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
x=np.linspace(0,10,100)
y=np.sin(x)

#Create a figure and an axis 
fig, ax = plt.subplots()

#Plotting 
ax.plot(x,y, label='sin(x)',color='blue')
ax.set_title('Sine function')
ax.set_xlabel('X values')
ax.set_ylabel('sin(x)')
ax.legend()
ax.grid(True)

#Display the plot
plt.show()

```

# Scatter plot

```python
np.random.seed(0)
x=np.random.rand(50)
y=np.random.rand(50)

fig,ax=plt.subplots()

ax.scatter(x,y,color='red')
ax.set_title('Random Scatter Plot')
ax.set_xlabel('X values')
ax.set_ylabel('Y values')
ax.grid(True)

#Display the plot
plt.show()
```

# Bar Cahr

```python
categories=['A','B','C','D']
values=[3,7,2,5]

fig, ax=plt.subplots()

ax.bar(categories,values,color='purple')
ax.set_title('Bar Plot Example')
ax.set_xlabel('Categories')
ax.set_ylabel('Values')

plt.show()
```

# Histogram

```python
data = np.random.randn(100)
fig, ax =plt .subplots()
ax.hist(data,bins=30, color='green',alpha=0.7)
ax.set_title('Histogram of random data')
ax.set_xlabel('Value')
ax.set_ylabel('Frequency')
ax.grid(True)

plt.show()
```

# Box Plot

```python
df = [np.random.normal(0,std,100) for std in range (1,4)]

fig, ax=plt.subplots()

ax.boxplot(df,vert=True,patch_artist=True)

ax.set_title("Box Plot Example")
ax.set_xlabel("Dataset")
ax.set_ylabel("Value")
ax.set_xticklabels(['Std 1','Std2','Std3'])

plt.show()
```
